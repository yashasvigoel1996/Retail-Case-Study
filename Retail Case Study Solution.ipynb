```python
# Solution 4(a). Time period of the available transaction data
import numpy as np

customer_trasaction_period=Customer_Final.copy()

date_formats = ['%d/%m/%Y', '%d-%m-%Y']

converted_transaction_dates=[]

for date in customer_trasaction_period.tran_date:
    for date_format in date_formats:
        try:
            converted_date=pd.to_datetime(date,format=date_format)
            converted_transaction_dates.append(converted_date)
        except ValueError:
            pass
        
customer_trasaction_period["converted_transaction_dates"]=converted_transaction_dates
customer_trasaction_period=customer_trasaction_period.sort_values(by=["converted_transaction_dates"],ascending=False)

latest_transaction_date=customer_trasaction_period.head(1).converted_transaction_dates.values[0]
latest_transaction_date=pd.to_datetime(latest_transaction_date,format="%d-%m-%Y")
oldest_transaction_date=customer_trasaction_period.tail(1).converted_transaction_dates.values[0]
oldest_transaction_date=pd.to_datetime(oldest_transaction_date,format="%d-%m-%Y")
timePeriod=(latest_transaction_date-oldest_transaction_date).days
print("Time period of the available transaction data is {} days".format(timePeriod))
```

```python
#6. Which City code has the maximum customers and what was the percentage of customers from that city?
#Solution -> City code has the maximum customers
city_maximum_customers=Customer_Final.groupby(by=["city_code"])[["city_code"]].count().add_prefix("Max_Of")
city_maximum_customers_max=city_maximum_customers.loc[(city_maximum_customers.Max_Ofcity_code.values==city_maximum_customers.max().values),]
city_maximum_customers_sum=city_maximum_customers.sum().add_prefix("Sum_Of_")
#city_maximum_customers_sum

percentage_of_customers=(city_maximum_customers_max.values[0]/city_maximum_customers_sum.values[0])*100

#city_maximum_customers
percentage_of_customers

#So City code 4.0 has maximum customers
#So 10.51% percent of customer from city code 4.0
```

```python
#Solutin 7. Which store type sells the maximum products by value and by quantity?

#store type sells the maximum products by value
store_type_max_value=Customer_Final.groupby(by=["Store_type"])[["total_amt"]].sum().add_prefix("Sum_Of_")
store_type_max_value=store_type_max_value.loc[(store_type_max_value.Sum_Of_total_amt.values==store_type_max_value.max().values),]
#store_type_max_value

#So, e-shop sells the maximum products by value

#store type sells the maximum products by quantity
store_type_max_quantity=Customer_Final.groupby(by=["Store_type"])[["Qty"]].sum().add_prefix("Sum_Of_")
store_type_max_quantity=store_type_max_quantity.loc[(store_type_max_quantity.Sum_Of_Qty.values==store_type_max_quantity.max().values),]
store_type_max_quantity

#So, e-shop sells the maximum products by quantity
```

```python
# Solution 11. For all customers aged between 25 - 35, find out:
# (a). What was the total amount spent for â€œElectronicsâ€ and â€œBooksâ€ product categories?
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta


customer_final_age=Customer_Final.copy()

date_formats = ['%d/%m/%Y', '%d-%m-%Y']
#testing_date=["14/04/1985","07-10-1982","12/02/1990","02-11-1989"]

#customer_final_age.DOB

converted_DOBs=[]
now = pd.Timestamp(datetime.now())
for date in customer_final_age.DOB:
    for date_format in date_formats:
        try:
            converted_date=pd.to_datetime(date,format=date_format)
            age = relativedelta(now, converted_date).years
            converted_DOBs.append(age)
        except ValueError:
            pass
        
customer_final_age["age"]=pd.Series(pd.to_numeric(converted_DOBs))

customer_final_age=customer_final_age.loc[(((customer_final_age["age"]>=25) & (customer_final_age["age"]<=35)) & ((customer_final_age["prod_cat"]=="Electronics") | (customer_final_age["prod_cat"]=="Books"))) ,]
```

```python
# solution b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?
total_spent_btw_dates=customer_final_age.copy()

date_formats = ['%d/%m/%Y', '%d-%m-%Y']

converted_transaction_dates=[]

for date in total_spent_btw_dates.tran_date:
    for date_format in date_formats:
        try:
            converted_date=pd.to_datetime(date,format=date_format)
            converted_transaction_dates.append(converted_date)
        except ValueError:
            pass
        
total_spent_btw_dates["converted_transaction_dates"]=converted_transaction_dates
start_date=pd.to_datetime("01 Jan,2014",format="%d %b,%Y")
end_date=pd.to_datetime("01 Mar,2014",format="%d %b,%Y")
total_spent_btw_dates=total_spent_btw_dates.loc[((total_spent_btw_dates.converted_transaction_dates>=start_date) & (total_spent_btw_dates.converted_transaction_dates<=end_date)),]
total_spent_btw_dates.total_amt.sum()
#So, 432298.1 (units) spent by these customers between 1st Jan, 2014 to 1st Mar, 2014.
```
